# Skewness Interpretation Table :

| Skewness Value   | Interpretation                      |
| ---------------- | ----------------------------------- |
| **0**            | Perfect symmetry                    |
| **0 to 0.5**    | Nearly symmetric (good)             |
| **0.5 to 1.0** | Moderate skew                       |
| **> 1**         | High skew (requires transformation) |


# Almost symmetric (good)

- Price  -0.03

- Quantity  -0.004

- Shipping Cost  -0.02

- Discount %  0.01

# Moderately right skewed 

- Total Amount  0.84 

# Highly right skewed

- Net Sales  1.03

# Mode

```python
df['Category']=df['Category'].fillna(df['Category'].mode()[0])
df['Category']
```

```python
df['Quantity']=df['Quantity'].fillna(df['Quantity'].mode()[0])
df['Quantity']
```

# transform() = "Do something group-wise, but give me back a column of the same size as original."

## Perfect for fillna, encoding, normalization, etc.

```python
df['Review_Text'] = (df.groupby('Category')['Review_Text'].transform(lambda x : x.fillna(x.mode()[0])))
```

# Mode is dangerous for CustomerID and ProductID
- IDs are meant to be unique

- They should not repeat

- Mode creates fake duplicates

- Analytical results become wrong


 Yes, mode will run

 But it is NOT correct for IDs

 IDs are not categorical

 You should NEVER fill IDs with mode in real analysis


# Placeholders are simple text values you put in place of missing IDs so that:

 No error happens

 No fake customers/products are created

 Data integrity stays correct

 You can continue your project safely

```python
df['CustomerID'] = df['CustomerID'].fillna("UnknownCustomer")
```

```python
df['ProductID'] = df['ProductID'].fillna("MissingCustomer")
```

# Categorical vs Categorical

```python
ct = pd.crosstab(df['Product_Category'], df['Review'])

# Grouped bar plot
plt.figure(figsize=(10, 6))
ct.plot(kind='bar', figsize=(10,6))  # each review = separate bar

plt.title("Product Category vs Review (Grouped Bar Plot)")
plt.xlabel("Product Category")
plt.ylabel("Review Count")
plt.xticks(rotation=45)
plt.legend(title="Review")
plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# remove placeholder customer
df_clean = df[df['CustomerID'] != "UnknownCustomer"]

# find top 10 real customers
top_customers = df_clean['CustomerID'].value_counts().head(7).index

ct = pd.crosstab(df_clean[df_clean['CustomerID'].isin(top_customers)]['CustomerID'],
                 df_clean['Product_Category'])

ct.plot(kind='bar', figsize=(12,6))

plt.title("Top 10 Customers vs Product Category (Cleaned)")
plt.ylabel("Count")
plt.xlabel("CustomerID")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
